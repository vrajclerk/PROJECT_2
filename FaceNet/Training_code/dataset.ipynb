```python
import cv2
import os

# Load the pre-trained Haar Cascade face detector
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Function to detect and crop faces
def detect_and_crop_face(image_path, output_size=(224, 224)):
    # Read the image
    img = cv2.imread(image_path)
    if img is None:
        print(f"Image {image_path} not found!")
        return None

    # Convert to grayscale for face detection
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Detect faces in the image
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    # If faces are detected, crop the first detected face
    if len(faces) > 0:
        (x, y, w, h) = faces[0]  # Take the first face
        face = img[y:y+h, x:x+w]  # Crop the face region
        # Resize the face to the desired output size
        face_resized = cv2.resize(face, output_size)
        face_gray = cv2.cvtColor(face_resized, cv2.COLOR_BGR2GRAY)
        return face_gray
    else:
        return None

# Function to remove an image
def remove_image(image_path):
    if os.path.exists(image_path):
        os.remove(image_path)
        print(f"Removed {image_path}")

# Folder structure
input_folder = './images'  # Base folder of the dataset

# Traverse through all subdirectories and process images
for root, dirs, files in os.walk(input_folder):
    for file in files:
        if file.endswith(('.jpg', '.jpeg', '.png')):  # Check for image files
            img_path = os.path.join(root, file)
            cropped_face = detect_and_crop_face(img_path)

            if cropped_face is not None:
                # Overwrite the original image with the resized face image
                cv2.imwrite(img_path, cropped_face)
                print(f"Processed {img_path}")
            else:
                # Remove the image if no face is detected
                remove_image(img_path)

print("Processing complete!")
```
